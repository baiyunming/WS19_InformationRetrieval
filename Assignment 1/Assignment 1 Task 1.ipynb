```python
import numpy as np
import math
from matplotlib import pyplot as plt

def TailDistribution (x):
    result = 1- math.pow(0.99,x)
    return result 

print(TailDistribution(2))
print(TailDistribution(3))
print(TailDistribution(500))
```

```python
p = np.arange(0,501)

result = np.zeros(501)
for i in range(0,501):
     result[i] = TailDistribution(i)
   
```

```python
plt.plot(p, result)
plt.xlabel('$P$')
plt.ylabel('$Probability$')
plt.title('Probability any of the components of Xp to lie outside of the interval [-2.576,2.576]')
plt.show()
```
